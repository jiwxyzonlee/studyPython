```python
# ì˜¤ëŠ˜ ë‚ ì”¨ë¥¼ ê°€ì ¸ì™€ì„œ ê·¸ë˜í”„ ê·¸ë¦¬ê¸°
import requests

# í•„ìš”í•œ html ê°€ì ¸ì˜¤ê¸°
resp = requests.get('https://www.weather.go.kr/weather/observation/currentweather.jsp')
text = resp.text
#print(text)

import bs4
bs = bs4.BeautifulSoup(text, 'html.parser')
#class ì†ì„±ì˜ ê°’ì´ table_develop3 ì¸ ë°ì´í„° ì°¾ì•„ì˜¤ê¸°
table = bs.select('.table_develop3')
#print(table)

# ë„ì‹œì´ë¦„, ì˜¨ë„, ìŠµë„ë¥¼ ì €ì¥í•  listë¥¼ ìƒì„±
loc = []
temp = []
hum = []

# ê°€ì ¸ì˜¨ í…Œì´ë¸” ì¤‘ì—ì„œ ì²«ë²ˆì§¸ í…Œì´ë¸”ë¡œë¶€í„° ì¤„(tr) ë‹¨ìœ„ë¡œ ì½ê¸°
for tr in table[0].find_all('tr'):
    tds = tr.find_all('td')  #tr íƒœê·¸ì—ì„œ tdë¥¼ ì „ë¶€ ì°¾ì•„ì„œ tdsì— ëŒ€ì…
    for td in tds:
        if td.find('a'): # a íƒœê·¸ê°€ ìˆìœ¼ë©´ ë„ì‹œ ì´ë¦„ì´ ìˆëŠ” í–‰ì…ë‹ˆë‹¤.
            # ë„ì‹œ ì´ë¦„ì„ locì— ì¶”ê°€
            loc.append(td.find('a').text)
            # ì˜¨ë„ë¥¼ tempì— ì¶”ê°€
            temp.append(tds[5].text)
            # ìŠµë„ë¥¼ tempì— ì¶”ê°€
            hum.append(tds[10].text)
            
# ë°ì´í„° í™•ì¸
# ì¸ë±ì‹±ì€ ë²ˆí˜¸ í•œ ê°œë§Œ ì‘ì„±í•´ì„œ í•˜ë‚˜ì˜ ë°ì´í„°ë§Œ ì¶”ì¶œ
#print(loc[0:5]) #0-4ë²ˆí–‰ê¹Œì§€ í™•ì¸ - ìŠ¬ë¼ì´ì‹±(ë²”ìœ„ë¥¼ ê°€ì§€ê³  ì¶”ì¶œ)
#print(temp[0:5])
#print(hum[0:5])

# í•„ìš”í•œ ë„ì‹œì˜ ë°ì´í„°ë§Œ ì¶”ì¶œ
cities = ['ì„œìš¸', 'ë¶€ì‚°', 'í¬í•­', 'ì¸ì²œ', 'ëŒ€êµ¬', 'ëŒ€ì „', 'ê´‘ì£¼', 'ìš¸ì‚°', 'ì°½ì›', 'í‘ì‚°ë„' ]
# ë„ì‹œë“¤ì˜ ì˜¨ë„ì™€ ìŠµë„ë¥¼ ì €ì¥í•  list
temperatures = []
humidities = []

# citiesì— ìˆëŠ” ë°ì´í„°ë§Œ ì¶”ì¶œí•´ì„œ ì €ì¥
for city in cities:
    j = 0
    for c in loc:
        if c == city:
            temperatures.append(temp[j])
            humidities.append(hum[j])
            break
        j = j + 1
        
#print('cities: ', cities)
#print('temperatures: ', temperatures)
#print('humidities: ', humidities)

####

# íŒŒì´ì¬ì—ì„œ ì‹œê°í™”ë¥¼ í•˜ëŠ” ê¸°ë³¸ íŒ¨í‚¤ì§€
import matplotlib.pyplot as plt

# í•œê¸€ í°íŠ¸ ì‚¬ìš©ì„ ìœ„í•œ íŒ¨í‚¤ì§€
from matplotlib import font_manager, rc, rcParams
import platform # ìš´ì˜ ì²´ì œ í™•ì¸ì„ ìœ„í•œ íŒ¨í‚¤ì§€

# ì£¼í”¼í„° ë…¸íŠ¸ë¶ì—ì„œ ì…€ ì•ˆì— ì¶œë ¥í•´ì£¼ëŠ” ì„¤ì •
# ì£¼í”¼í„° ë…¸íŠ¸ë¶ì—ì„œ ê·¸ë˜í”„ê°€ í´ë˜ìŠ¤ ì´ë¦„ìœ¼ë¡œ ì¶œë ¥ - í•œë²ˆ ë” ì‹¤í–‰í•´ë„ ë¨
%matplotlib inline

# ìŒìˆ˜ë¥¼ ì œëŒ€ë¡œ í‘œí˜„í•˜ê¸° ìœ„í•œ ì„¤ì •
rcParams['axes.unicode_minus'] = False
# ì´ ì„¤ì •ì„ í•˜ì§€ì•Šìœ¼ë©´ ìŒìˆ˜ê°€ ë„¤ëª¨ë¡œ ì¶œë ¥

# í•œê¸€ í°íŠ¸ ì‹¤í–‰
if platform.system() == 'Windows': # ìœˆë„ìš°ì¦ˆ
    font_name = font_manager.FontProperties(fname='c:/windows/Fonts/malgun.ttf').get_name()
    rc('font', family=font_name)
elif platform.system() == 'Darwin': #ë§¤í‚¨í† ì‹œ
    rc('font', family='AppleGothic')
else: 
    print('ì•Œ ìˆ˜ ì—†ëŠ” ìš´ì˜ì²´ì œ')
 

# ê·¸ë˜í”„ ê·¸ë¦¬ê¸°
plt.figure(figsize=(12, 4), dpi = 300)
#ê·¸ë˜í”„ì˜ í¬ê¸° ì„¤ì • - ê°€ë¡œ ì„¸ë¡œ 12, 4ì¸ì¹˜

# êº¾ì€ì„  ê·¸ë˜í”„ - plot
plt.plot(temperatures, label = 'ì˜¨ë„', lw=3, color='r', linestyle='-', marker='s')
plt.plot(humidities, label = 'ìŠµë„', lw=3, color='g', linestyle='-', marker='s')

# xì¶•
plt.xticks(range(0, len(cities), 1), cities, rotation='vertical')

# ë²”ë¡€
plt.legend()

# ê·¸ë¦¼íŒŒì¼ë¡œ ì €ì¥
plt.savefig('graph.png', dpi = 300)

plt.show()
```
